# 평균 주택 가격을 연령별로 비교

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('/content/sample_data/california_housing_test.csv')

plt.figure(figsize=(10, 5))
df.groupby("housing_median_age")["median_house_value"].mean().plot(kind="line", marker="o", color="b")
plt.xlabel("Housing Median Age")
plt.ylabel("Average Median House Value")
plt.title("Average House Value by Housing Median Age")
plt.grid(True)
plt.show()
```

# 소득 구간별 주택 가격 분포

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('/content/sample_data/california_housing_test.csv')

plt.figure(figsize=(12, 6))
sb.boxplot(x=pd.qcut(df["median_income"], q=5), y=df["median_house_value"])
plt.xlabel("Median Income Groups")
plt.ylabel("Median House Value")
plt.title("House Price Distribution by Income Groups")
plt.xticks(rotation=45)
plt.show()
```
